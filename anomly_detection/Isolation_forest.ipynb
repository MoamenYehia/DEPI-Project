```python
df_model = df.copy()
```

```python
drop_cols = [
    "order_id",
    "customer_id",
    "customer_unique_id",
    "product_id",
    "seller_id",
    "review_id",
    "review_comment_title",
    "review_comment_message"
]

df_model = df_model.drop(columns=drop_cols)
```

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "shipping_limit_date",
    "review_creation_date",
    "review_answer_timestamp"
]

for col in date_cols:
    df_model[col] = pd.to_datetime(df_model[col], errors="coerce")
```

```python
df_model["purchase_year"] = df_model["order_purchase_timestamp"].dt.year
df_model["purchase_month"] = df_model["order_purchase_timestamp"].dt.month
df_model["purchase_day"] = df_model["order_purchase_timestamp"].dt.day
df_model["purchase_weekday"] = df_model["order_purchase_timestamp"].dt.weekday
df_model["purchase_hour"] = df_model["order_purchase_timestamp"].dt.hour
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers

    lat1, lon1, lat2, lon2 = map(
        radians,
        [lat1, lon1, lat2, lon2]
    )

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        sin(dlat / 2) ** 2
        + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    )

    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c
```

```python
df_model["distance_km"] = df_model.apply(
    lambda row: haversine(
        row["customer_lat"],
        row["customer_lng"],
        row["seller_lat"],
        row["seller_lng"]
    ),
    axis=1
)
```

```python
df_model = df_model.drop(columns=date_cols)
```

```python
drop_more_cols = [
    "customer_city",
    "seller_city",
    "customer_zip_code_prefix",
    "seller_zip_code_prefix",
    "customer_lat",
    "customer_lng",
    "seller_lat",
    "seller_lng"
]

df_model = df_model.drop(columns=drop_more_cols)
```

```python
df_model.isnull().sum().sort_values(ascending=False)
```

```python
num_cols = df_model.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols:
    df_model[col] = df_model[col].fillna(df_model[col].median())
```

```python
cat_cols = df_model.select_dtypes(include="object").columns

for col in cat_cols:
    df_model[col] = df_model[col].fillna("Unknown")
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}

cat_cols = df_model.select_dtypes(include="object").columns

for col in cat_cols:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col].astype(str))
    label_encoders[col] = le
```

```python
df_model["total_cost"] = df_model["price"] + df_model["freight_value"]
```

```python
df_model["payment_per_installment"] = (
    df_model["payment_value"] / df_model["payment_installments"]
)
```

```python
df_model["payment_installments"] = df_model["payment_installments"].replace(0, 1)
df_model["payment_per_installment"] = (
    df_model["payment_value"] / df_model["payment_installments"]
)
```

```python
df_model["freight_ratio"] = (
    df_model["freight_value"] / (df_model["price"] + 1)
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(df_model)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)

model.fit(X)
```

```python
predictions = model.predict(X)
```

```python
df["anomaly"] = predictions
```

```python
def predict_anomaly(new_data):
    """
    new_data should be a DataFrame with the same columns
    used during training.
    """

    new_scaled = scaler.transform(new_data)

    prediction = model.predict(new_scaled)

    return prediction
```
